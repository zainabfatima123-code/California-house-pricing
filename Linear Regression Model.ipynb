**PREPARING THE DATASET**

```python
dataset=pd.DataFrame(california.data,columns=california.feature_names)
```

```python
dataset['Price']=california.target
```

```python
## independent and dependent features
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
##3 internally in linear regression we use gradient descent our main aim is to reach the global minima and to reach to the global minima we need to make sure that all independent feature units are same so that we reach the global minima as soon as possible
### Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
## pass x tarin dataset and transform it completely
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

**MODEL TRAINING**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#### print the coefficient and the intercept
print(regression.coef_)
## it would be equal to number of independent features
```

```python
print(regression.intercept_)
```

```python
##3 on which parameters the model has been traine
regression.get_params()
```

```python
#### prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
### if the plot is linear it means ur model is performng well
```

```python

residuals=y_test-reg_pred
#### residuals
```

```python
residuals
```

```python
### plotting this residuals
sns.displot(residuals,kind='kde')
```

```python
##3 scatter plot with respect to prediction and residuals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

 **r square and adjusted r square**

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
### it should be more towards 1
```

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
